```python
# import markdown
# Iterate over each table in the document
for table in nqld.tables:
    # Store the table data in a list of lists
    table_data = []
    for row in table.rows:
        row_data = [cell.text for cell in row.cells]
        table_data.append(row_data)

    # Convert the table data to Markdown format
    table_md = ""
    for row_data in table_data:
        table_md += "| {} |\n".format(" | ".join(row_data))
    # html_text=markdown.markdown(table_md)
    # Print the Markdown table
    print(table_md)
```

## text in the doc

```python
# store the paragraph objects into a dataframe for easily manipulation later. 
df=pd.DataFrame(nqld.paragraphs, columns=["para_obj"])
df.head(3)
```

```python
df['text']=[i.text.strip() for i in df.para_obj]
df['length']=df.text.str.len()
df['style_']=[i.style.name for i in df.para_obj]
df=df.query('length>0').reset_index(drop=True).drop(columns='para_obj')
df
```

```python
df.query('style_.str.contains("Heading")').style_.value_counts().sort_index()
```

```python
df['h1']=None
df['h2']=None
df['h3']=None

h1_rows = df.query("style_=='Heading 1'").index
df.loc[h1_rows, 'h1'] = df.loc[h1_rows, 'text']

h2_rows = df.query("style_=='Heading 2'").index
df.loc[h2_rows, 'h2'] = df.loc[h2_rows, 'text']

h3_rows = df.query("style_=='Heading 3'").index
df.loc[h3_rows, 'h3'] = df.loc[h3_rows, 'text']

df.query('style_.str.contains("Heading")')
```

```python
df.h1.ffill(inplace=True)
df.h2=df.groupby('h1')['h2'].transform(lambda x: x.ffill())
df.h3=df.groupby(['h1', 'h2'])['h3'].transform(lambda x: x.ffill())
```

```python
df.to_excel('../data/interim/nqld.xlsx')
```

```python
df[['h1', 'h2', 'h3']]=df[['h1', 'h2', 'h3']].fillna("-")
df['heading'] = df.h1 + "/" + df.h2 + "/" + df.h3
df.heading=df.heading.str.replace("-/-/-", '').str.replace("/-", "").str.replace("-/", "")
df
```

```python
for i in df.query('heading!=""').heading:
    print(len(i), "    ",i)
```

```python
df=df.drop(index=df.query('style_.str.contains("Heading")').index, columns=['h1','h2','h3'])
```

```python
df=df[['heading', 'text']].copy()
```

```python
# Assuming your DataFrame is called 'df' and you want to store the result in a new DataFrame called 'grouped_df'
grouped_df = df.groupby('heading', sort=False)['text'].agg('\n'.join).reset_index()
grouped_df
```

```python
grouped_df['len']=grouped_df.text.str.len()
```

```python
grouped_df.style.background_gradient(cmap='twilight', subset='len', vmin=0, vmax=1000)
```

```python
grouped_df.text=grouped_df.text.str.replace(";\n", '.\n') # modify this to better cut those bulleting paragraphs with no "." in the doc. 
```

```python
text_splitter = ts.CharacterTextSplitter(
        separator=".",
        chunk_size=600,
        chunk_overlap=100,
        length_function=len
    )
text_splitter2 = ts.CharacterTextSplitter(
        separator=",",
        chunk_size=600,
        chunk_overlap=100,
        length_function=len
    )
```

```python
grouped_df.loc[grouped_df['text'].str.len() > 800, 'text'] = grouped_df['text'].apply(lambda x: text_splitter.split_text(x))
# Expanding the list into new rows
grouped_df = grouped_df.explode('text')
grouped_df.reset_index(drop=True, inplace=True) # if I don't do this, executing the second time (below) will get this error: "cannot reindex on an axis with duplicate labels"
```

```python


# reapply it the second time to eleminate all large size text_chunks
grouped_df.loc[grouped_df['text'].str.len() > 800, 'text'] = grouped_df['text'].apply(lambda x: text_splitter2.split_text(x))
# Expanding the list into new rows
grouped_df = grouped_df.explode('text')
grouped_df.reset_index(drop=True, inplace=True)
```

```python
grouped_df.drop(columns='len', inplace=True)
```

```python
grouped_df['len2']=grouped_df.text.str.len()
```

```python
grouped_df.style.background_gradient(cmap='Reds', subset='len2', vmin=1000, vmax=1010)
```
